```python
try:
    from PIL import Image
except ImportError:
    import Image

from pytesseract import pytesseract

import os
import numpy as np
import cv2
import datetime
import io
from PyPDF2 import PdfMerger

try:
    from tkinter import Tk
    from tkFileDialog import askopenfilenames
except:
    from tkinter import Tk
    from tkinter import filedialog

root = Tk()
root.withdraw()
root.call('wm', 'attributes', '.', '-topmost', True)
```

# SETTINGS

```python
# path_to_tesseract = "C:/Program Files/Tesseract-OCR/tesseract.exe"
# pytesseract.tesseract_cmd = path_to_tesseract
merger = PdfMerger()

# video_path = 'C:/Users/<USERNAME>/Desktop/<VIDEO_NAME>'
video_path = filedialog.askopenfilename()
video_name = os.path.basename(video_path).split('.')[0]


cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)

# From which frame to start scaning
# start_frame = fps * (60 * 0 + 20) # 8 min 42 sec
start_frame = 0

# Which frame will be shown in the preview
preview_frame = fps * 60 * 10
fps
```

# Preview (set cropping parameters and check them in the preview)

```python
# Some cropping parameters i used

# Full screen
# top_left_corner = [
#     (0,0)
# ]

# bottom_right_corner = [
#     (1920,1080)
# ]

# Standart
top_left_corner = [
    (134,73)
]

bottom_right_corner = [
    (1785,1000)
]

# Half screen
# top_left_corner = [
#     (18,273)
# ]

# bottom_right_corner = [
#     (952,800)
# ]

# With the participants shown
# top_left_corner = [
#     (16,152)
# ]

# bottom_right_corner = [
#     (1377,918)
# ]

# Misc
# top_left_corner = [
#     (340,73)
# ]

# bottom_right_corner = [
#     (1579,999)
# ]
```

```python
# Preview window
cap.set(cv2.CAP_PROP_POS_FRAMES, preview_frame)
# Read Images
ret, frame = cap.read()

if top_left_corner and bottom_right_corner:
    
    frame = frame[top_left_corner[0][1]:bottom_right_corner[0][1],top_left_corner[0][0]:bottom_right_corner[0][0]]
    cv2.imwrite("Preview.jpg", frame)
    cv2.imshow("Preview", frame)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

# Create images

```python
try:
    # creating a folder named data
    if not os.path.exists('images_'+video_name):
        os.makedirs('images_'+video_name)
except OSError:
    print ('Error: Creating directory of data')
```

```python
currentframe = start_frame
count = 0

difference = 1
prev_frame = None

cap.set(cv2.CAP_PROP_POS_FRAMES, currentframe)
while cap.isOpened():
    
    # reading from frame
    ret,frame = cap.read()
  
    if ret:
        name = f'./images_{video_name}/frame{str(count)}.jpg'
        
        frame = frame[top_left_corner[0][1]:bottom_right_corner[0][1],top_left_corner[0][0]:bottom_right_corner[0][0]]
        
        ret, bin_frame = cv2.threshold(frame,127,255,cv2.THRESH_BINARY)
        
        if prev_frame is not None:
            difference = np.mean(prev_frame != bin_frame)
            
        prev_frame = bin_frame
        
        if difference > 0.05:
            cv2.imwrite(name, frame)
            #print(f"Slide {count}:", str(datetime.timedelta(seconds=currentframe/60)))
            count += 1
        else:
            
            #print("same as 10 sec ago:", str(datetime.timedelta(seconds=currentframe/60)))
        
        
        currentframe += fps * 10 # 60 frames * 10 = 10 seconds
        cap.set(cv2.CAP_PROP_POS_FRAMES, currentframe)
    else:
        cap.release()
        break
        
print("FIN")
```

# Images to pdf

```python
for i in range(0,count):
    image_path = f'./images_{video_name}/frame{str(i)}.jpg'
#     print(image_path)
    img = Image.open(image_path)
    
    pdf = pytesseract.image_to_pdf_or_hocr(img, extension='pdf', lang='ukr+eng')
    
    bytes_pdf = io.BytesIO(pdf)
    
    merger.append(bytes_pdf)
        
merger.write(video_name + "_pdf.pdf")
merger.close()
print("FIN")
```
